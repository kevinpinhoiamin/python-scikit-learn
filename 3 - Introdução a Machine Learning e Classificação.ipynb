```python
import pandas as pd

uri = "https://gist.githubusercontent.com/guilhermesilveira/1b7d5475863c15f484ac495bd70975cf/raw/16aff7a0aee67e7c100a2a48b676a2d2d142f646/projects.csv"
dados = pd.read_csv(uri)
dados.head()
```

```python
a_renomear = {
    'expected_hours': 'horas_esperadas',
    'price': 'preco',
    'unfinished': 'nao_finalizado'
}
dados.rename(columns=a_renomear, inplace=True)
dados.head()
```

```python
troca = {
    0: 1,
    1: 0
}
dados['finalizado'] = dados.nao_finalizado.map(troca)
dados.head()
```

```python
dados.tail()
```

```python
import seaborn as sns
%matplotlib inline
sns.set(rc={'figure.figsize':(12, 6)})

sns.scatterplot(x='horas_esperadas', y='preco', data=dados)
```

```python
sns.scatterplot(x='horas_esperadas', y='preco', hue='finalizado', data=dados)
```

```python
sns.relplot(x='horas_esperadas', y='preco', hue='finalizado', col='finalizado', data=dados)
```

```python
x = dados[['horas_esperadas', 'preco']]
y = dados['finalizado']
```

```python
from sklearn.model_selection import train_test_split
from sklearn.svm import LinearSVC
from sklearn.metrics import accuracy_score

SEED = 5

treino_x, teste_x, treino_y, teste_y = train_test_split(x, y, random_state=SEED, test_size=0.25, stratify=y)
print(f'Treinaremos com {len(treino_x)} elementos e testaremos com {len(teste_x)} elementos.')

modelo = LinearSVC()
modelo.fit(treino_x, treino_y)
previsoes = modelo.predict(teste_x)

acuracia = (accuracy_score(teste_y, previsoes) * 100).round(2)
print(f'A acurácia foi de {acuracia}%.')
```

```python
import numpy as np

previsoes_de_base = np.ones(540)

acuracia = (accuracy_score(teste_y, previsoes_de_base) * 100).round(2)
print(f'A acurácia do algoritmo de previsão de baseline foi de {acuracia}%.')
```

```python
sns.scatterplot(x='horas_esperadas', y='preco', hue=teste_y, data=teste_x)
```

```python
x_min = teste_x.horas_esperadas.min()
x_max = teste_x.horas_esperadas.max()
y_min = teste_x.preco.min()
y_max = teste_x.preco.max()
print(x_min, x_max, y_min, y_max)
```

```python
pixels = 100
eixo_x = np.arange(x_min, x_max, (x_max - x_min) / pixels)
eixo_y = np.arange(y_min, y_max, (y_max - y_min) / pixels)
```

```python
xx, yy = np.meshgrid(eixo_x, eixo_y)
pontos = np.c_[xx.ravel(), yy.ravel()]
pontos
```

```python
Z = modelo.predict(pontos)
Z = Z.reshape(xx.shape)
Z
```

```python
import matplotlib.pyplot as plt

plt.contourf(xx, yy, Z, alpha=0.3)
plt.scatter(teste_x.horas_esperadas, teste_x.preco, c=teste_y, s=4)

# DECISION BOUNDARY
```
